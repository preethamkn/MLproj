```python
import numpy as np
import pandas as pd
from sklearn import metrics
from pandas.plotting import scatter_matrix 
import matplotlib.pyplot as plt 
from sklearn import model_selection 
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train = pd.read_csv("Computers2.csv")
train.describe()
```

```python
feature_names = ['speed','hd','ram','screen','cd','premium']
```

```python
data = pd.DataFrame()
```

```python
prem={'yes':1,'no':0}
cd_c={'yes':1,'no':0}
def clean_prem(x):
    return prem[x]
def clean_cd_c(x):
    return cd_c[x]
data['price']=train['price']
data['speed']=train['speed']
data['hd']=train['hd']
data['ram']=train['ram']
data['screen']=train['screen']
data['cd']=train['cd'].apply(clean_cd_c)
data['premium']=train['premium'].apply(clean_prem)
```

```python
data.head() 
```

```python
X = data[feature_names]
X
```

```python
y = data.price
y.values.ravel()
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state=20)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import r2_score
model=ExtraTreesRegressor()
model.fit(X_train,y_train)
prediction = model.predict(X_test)
r2score=eval("%.2f" % r2_score(y_test,prediction))
rmse=np.sqrt(mean_squared_error(y_test,prediction))
```
